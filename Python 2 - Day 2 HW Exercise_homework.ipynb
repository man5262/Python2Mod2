## Exercise 1: 
Create a Temperature Converter program using OOP by creating a Temperature class with two private attributes to store Fahrenheit and Celsius degrees. In the Temperature class, define methods that 
- sets the private attributes. When you set one unit of temperature, it should calculate and set the other unit of temperature. For example, when you set degrees in Fahrenheit, it should calculate and set in Celsius degrees. 
- gets the hidden attributes that round the number to 2 decimal places. 

The output should look something like following:

    MENU
    1. Fahrenheit to Celsius
    2. Celsius to Fahrenheit
    3. Quit
    
    Enter a menu option: 1
    Enter degrees in Fahrenheit: 99
    99.00 oF is 37.22 oC.
    
    Enter a menu option: 2
    Enter degrees in Celsius: 37.22
    37.22 oC is 99.00 oF.
    
    Enter a menu option: 3
    Bye


```python
class Temperature:

    def __init__(self):
#        self.__source_degree = degree
        self.__fahrenheit_temp = 0
        self.__celsius_temp = 0

    # Set Fahrenheit temperature
    def set_fahrenheit_temp(self, temp):
        self.__fahrenheit_temp = temp

    # Set Celsius temperature
    def set_celsius_temp(self, temp):
        self.__celsius_temp = temp

    # Get Fahrenheit temperature
    def get_fahrenheit_temp(self):
        return self.__fahrenheit_temp

    # Get Celsius temperature
    def get_celsius_temp(self):
        return self.__celsius_temp
    
    # Temperature converter method:
    # If degree = 1 then convert from Fahrenheit to Celsius
    # If degree = 2 then convert from Celsius to Fahrenheit
    def temperature_converter(self, degree, temp):
        if degree == 1:
            self.set_fahrenheit_temp(temp)
            self.set_celsius_temp((self.__fahrenheit_temp - 32)   * 5 / 9)           
        elif degree == 2:
            self.set_celsius_temp(temp)
            self.set_fahrenheit_temp(self.__celsius_temp * 9 / 5 + 32)           

def display_menu():
    menu = ["1. Fahrenheit to Celsius",
            "2. Celsius to Fahrenheit",
            "3. Quit"]

    print ("MENU")
    for menu_item in menu:
        print (menu_item)


degrees = ["Fahrenheit", "Celsius"]
temps = [0.0, 0.0]

# Display menu and select an option
while True:
    display_menu()
    opt_str = input("Enter a menu option: ")

    try:

        # Check an option and convert temperature accordingly
        opt = int(opt_str)
#        print(opt)
        if (opt == 3):
            print("\nBye.")
            break
        if opt != 1 and opt != 2:
            continue
    except:
        print(F"ERROR: {opt_str} is not a number\n")
        continue

    t = Temperature()

    # Enter temperature and convert it
    temp = float(input(f"Enter degrees in {degrees[opt - 1]}: "))
    t.temperature_converter(opt, temp)
    
    #Print result
    temps[0] = t.get_fahrenheit_temp()
    temps[1] = t.get_celsius_temp()
    print(f"{temps[opt - 1]} o{degrees[opt - 1][0]} is {temps[opt - 2]:.2f} o{degrees[opt - 2][0]}\n")
```

```python
# From Day 1 Class Demo
class Product : 
    """A simple attempt to model a product."""
    
    def __init__(self, name, price, discount_rate) :
        """Initialize name, price, and discount_rate attributes"""
        self.__name = name
        self.__price = price
        self.__discount_rate = discount_rate
        
        
    def get_discount_amount(self) :
        """Computes a discount calculation"""
        return self.__price * self.__discount_rate / 100
    
        
    def get_info(self) :
        return(f"Name: {self.__name}\n" +
                f"Price: {self.__price}\n" +
                f"Disicount Amount: ${self.get_discount_amount():,.2f}\n" +
                f"Discounted Price: ${self.get_sale_price():,.2f}")
    
    def get_sale_price(self) :
        """Calls another method to find a sale price"""
        return self.__price - self.get_discount_amount()
    
    def get_description(self):
        return "There is no description at this time"
    
class Book(Product):
    def __init__(self, author):
        super().__init__("Gone with the Wind", 60, 25)
        self.__author = author
    
    def get_author(self):
        return self.__author
    
    def get_description(self):
        return (f"This book is written by {self.get_author()}")
    
class Movie(Product):
    def __init__(self, year):
        super().__init__("Snow White", 30, 15)
        self.__year = year
    
    def get_year(self):
        return self.__year
    
    def get_description(self):
        return (f"This movie was released in {self.get_year()}")

# Print information from Product object
print("Print information about Product object\n")
p = Product("Bob", 15.67, 20)
print(f"{p.get_info()}")
print(f"Description: {p.get_description()}\n\n")

# Print information from Book object
print("Print information about Book object\n")
b = Book("Margaret Mitchell")
print(f"{b.get_info()}")
print(f"Description: {b.get_description()}")

# Print information from Movie object
print("Print information about Movie object\n")
m = Movie(1938)
print(f"{m.get_info()}")
print(f"Description: {m.get_description()}")
```
